```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
iwp_df = pd.read_csv('/content/IWP.csv')
iwp_df = iwp_df.drop(index=0)
iwp_df['Date'] = pd.to_datetime(iwp_df['Date'])
iwp_df.set_index('Date', inplace=True)

# Convert necessary columns
numeric_cols = ['Open', 'High', 'Low', 'Close', 'Volume', 'Return', 'Short_MA', 'Long_MA']
iwp_df[numeric_cols] = iwp_df[numeric_cols].astype(float)

# Calculate 50-day and 200-day MAs
iwp_df['MA_50'] = iwp_df['Close'].rolling(50).mean()
iwp_df['MA_200'] = iwp_df['Close'].rolling(200).mean()

# Compute daily & cumulative return
iwp_df['Daily_Return'] = iwp_df['Close'].pct_change().fillna(0)
iwp_df['Cumulative_Return'] = (1 + iwp_df['Daily_Return']).cumprod()

# Signal: 1 when MA_50 > MA_200 (golden cross), -1 when MA_50 < MA_200 (death cross)
iwp_df['Crossover_Signal'] = 0
iwp_df.loc[iwp_df['MA_50'] > iwp_df['MA_200'], 'Crossover_Signal'] = 1
iwp_df.loc[iwp_df['MA_50'] < iwp_df['MA_200'], 'Crossover_Signal'] = -1
iwp_df['Signal_Change'] = iwp_df['Crossover_Signal'].diff()

# Extract signal dates
golden_cross = iwp_df[iwp_df['Signal_Change'] == 2]
death_cross = iwp_df[iwp_df['Signal_Change'] == -2]

# Plot
plt.figure(figsize=(16, 6))
plt.plot(iwp_df.index, iwp_df['Cumulative_Return'], label='Cumulative Return', alpha=0.8)
plt.scatter(golden_cross.index, golden_cross['Cumulative_Return'], color='green', label='Golden Cross (Buy)', marker='^', s=100)
plt.scatter(death_cross.index, death_cross['Cumulative_Return'], color='red', label='Death Cross (Sell)', marker='v', s=100)
plt.title("IWP: Cumulative Return with Golden/Death Cross Strategy (50 vs 200 MA)")
plt.xlabel("Date")
plt.ylabel("Cumulative Return")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load and clean data
iws_df = pd.read_csv('/content/IWS.csv')
iws_df = iws_df.drop(index=0)
iws_df['Date'] = pd.to_datetime(iws_df['Date'])
iws_df.set_index('Date', inplace=True)

# Convert numeric columns
numeric_cols = ['Open', 'High', 'Low', 'Close', 'Volume', 'Return', 'Short_MA', 'Long_MA']
iws_df[numeric_cols] = iws_df[numeric_cols].astype(float)

# Calculate moving averages
iws_df['MA_50'] = iws_df['Close'].rolling(50).mean()
iws_df['MA_200'] = iws_df['Close'].rolling(200).mean()

# Daily return
iws_df['Daily_Return'] = iws_df['Close'].pct_change().fillna(0)

# Buy & Hold cumulative return
iws_df['BuyHold_Cumulative'] = (1 + iws_df['Daily_Return']).cumprod()

# Golden/Death cross strategy
iws_df['Signal'] = 0
iws_df.loc[iws_df['MA_50'] > iws_df['MA_200'], 'Signal'] = 1

# Strategy return: invest only when signal is active
iws_df['Strategy_Return'] = iws_df['Daily_Return'] * iws_df['Signal']
iws_df['Strategy_Cumulative'] = (1 + iws_df['Strategy_Return']).cumprod()

# Plot both
plt.figure(figsize=(16, 6))
plt.plot(iws_df.index, iws_df['BuyHold_Cumulative'], label='Buy & Hold', linestyle='--')
plt.plot(iws_df.index, iws_df['Strategy_Cumulative'], label='Golden/Death Cross Strategy', linestyle='-')
plt.title("IWS: Strategy vs Buy & Hold (50-day vs 200-day MA)")
plt.xlabel("Date")
plt.ylabel("Cumulative Return")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load and clean IWP data
iwp_df = pd.read_csv('/content/IWP.csv').drop(index=0)
iwp_df['Date'] = pd.to_datetime(iwp_df['Date'])
iwp_df.set_index('Date', inplace=True)
iwp_df[['Close']] = iwp_df[['Close']].astype(float)
iwp_df['Daily_Return'] = iwp_df['Close'].pct_change().fillna(0)
iwp_df['Cumulative_Return'] = (1 + iwp_df['Daily_Return']).cumprod()
iwp_df['Cumulative_Max'] = iwp_df['Cumulative_Return'].cummax()
iwp_df['Drawdown'] = iwp_df['Cumulative_Return'] / iwp_df['Cumulative_Max'] - 1

# Repeat for IWS
iws_df = pd.read_csv('/content/IWS.csv').drop(index=0)
iws_df['Date'] = pd.to_datetime(iws_df['Date'])
iws_df.set_index('Date', inplace=True)
iws_df[['Close']] = iws_df[['Close']].astype(float)
iws_df['Daily_Return'] = iws_df['Close'].pct_change().fillna(0)
iws_df['Cumulative_Return'] = (1 + iws_df['Daily_Return']).cumprod()
iws_df['Cumulative_Max'] = iws_df['Cumulative_Return'].cummax()
iws_df['Drawdown'] = iws_df['Cumulative_Return'] / iws_df['Cumulative_Max'] - 1

# Plot drawdown timelines
plt.figure(figsize=(16, 6))
plt.plot(iwp_df.index, iwp_df['Drawdown'], label='IWP Drawdown', color='red')
plt.plot(iws_df.index, iws_df['Drawdown'], label='IWS Drawdown', color='blue')
plt.title("Rolling Drawdown Timeline: IWP vs IWS")
plt.xlabel("Date")
plt.ylabel("Drawdown (%)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plotting the cumulative return comparison as a bar chart (last value per year)

# Extract last cumulative return for each year
iwp_yearly_cum_return = iwp_df['Cumulative_Return'].resample('Y').last()
iws_yearly_cum_return = iws_df['Cumulative_Return'].resample('Y').last()

# Combine into a single DataFrame
yearly_cum_returns = pd.DataFrame({
    'IWP': iwp_yearly_cum_return,
    'IWS': iws_yearly_cum_return
})

# Convert index to year for labeling
yearly_cum_returns.index = yearly_cum_returns.index.year

# Plotting
plt.figure(figsize=(16, 6))
width = 0.35
x = yearly_cum_returns.index

plt.bar(x - 0.175, yearly_cum_returns['IWP'], width=width, label='IWP (Growth)')
plt.bar(x + 0.175, yearly_cum_returns['IWS'], width=width, label='IWS (Value)')

plt.title("Year-End Cumulative Return Comparison: IWP vs IWS")
plt.xlabel("Year")
plt.ylabel("Cumulative Return")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True, axis='y')
plt.tight_layout()
plt.show()
```

#####    IWP (blue bars) often outperforms in bull markets but fluctuates more.

IWS (yellow bars) offers steadier returns, especially in volatile years.

```python
# Extract yearly maximum drawdown for each ETF
iwp_yearly_drawdown = iwp_df['Drawdown'].resample('Y').min() * 100  # Convert to %
iws_yearly_drawdown = iws_df['Drawdown'].resample('Y').min() * 100  # Convert to %

# Combine into a single DataFrame
yearly_drawdown = pd.DataFrame({
    'IWP': iwp_yearly_drawdown,
    'IWS': iws_yearly_drawdown
})

# Convert index to year
yearly_drawdown.index = yearly_drawdown.index.year

# Plotting drawdown comparison as bar chart
plt.figure(figsize=(16, 6))
x = yearly_drawdown.index
width = 0.35

plt.bar(x - 0.175, yearly_drawdown['IWP'], width=width, label='IWP (Growth)', color='blue')
plt.bar(x + 0.175, yearly_drawdown['IWS'], width=width, label='IWS (Value)', color='orange')

plt.title("Yearly Maximum Drawdown Comparison: IWP vs IWS")
plt.xlabel("Year")
plt.ylabel("Max Drawdown (%)")
plt.xticks(rotation=45)
plt.grid(True, axis='y')
plt.legend()
plt.tight_layout()
plt.show()
```

#### insights

Drawdown represents the biggest drop from peak to trough in each year.

IWP (blue) frequently experiences larger drawdowns, reflecting higher volatility.

IWS (orange) tends to have milder declines, indicating greater stability.

In years like 2008 and 2020, both ETFs suffered significant drawdowns due to market crashes.

But typically, Value (IWS) protects capital better in downturns than Growth (IWP).

```python
import matplotlib.pyplot as plt

# Calculate rolling 30-day volatility (standard deviation of daily returns)
iwp_df['30D_Volatility'] = iwp_df['Return'].rolling(window=30).std()
iws_df['30D_Volatility'] = iws_df['Return'].rolling(window=30).std()

# Calculate cumulative returns
iwp_df['Cumulative_Return'] = (1 + iwp_df['Return'].fillna(0)).cumprod()
iws_df['Cumulative_Return'] = (1 + iws_df['Return'].fillna(0)).cumprod()

# Calculate average monthly close prices
iwp_monthly = iwp_df['Close'].resample('M').mean()
iws_monthly = iws_df['Close'].resample('M').mean()

# Prepare drawdown
iwp_df['Cumulative_Max'] = iwp_df['Cumulative_Return'].cummax()
iwp_df['Drawdown'] = iwp_df['Cumulative_Return'] / iwp_df['Cumulative_Max'] - 1

iws_df['Cumulative_Max'] = iws_df['Cumulative_Return'].cummax()
iws_df['Drawdown'] = iws_df['Cumulative_Return'] / iws_df['Cumulative_Max'] - 1

# Plotting
plt.figure(figsize=(14, 6))
plt.plot(iwp_df['Cumulative_Return'], label='IWP - Growth')
plt.plot(iws_df['Cumulative_Return'], label='IWS - Value')
plt.title("Cumulative Return Comparison")
plt.ylabel("Cumulative Return")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(14, 6))
plt.plot(iwp_df['30D_Volatility'], label='IWP - Growth')
plt.plot(iws_df['30D_Volatility'], label='IWS - Value')
plt.title("Rolling 30-Day Volatility")
plt.ylabel("Volatility (Std Dev of Return)")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(14, 6))
plt.plot(iwp_df['Drawdown'], label='IWP - Growth')
plt.plot(iws_df['Drawdown'], label='IWS - Value')
plt.title("Drawdown Comparison")
plt.ylabel("Drawdown")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(14, 6))
plt.plot(iwp_monthly, label='IWP - Growth')
plt.plot(iws_monthly, label='IWS - Value')
plt.title("Average Monthly Close Price")
plt.ylabel("Average Close Price")
plt.legend()
plt.grid(True)
plt.show()
```
